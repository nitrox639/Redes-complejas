```python
artistas = []
for url in urls:
    response = requests.get(url)
    soup = BS(response.content)
    elements = soup.findAll(attrs = {"class": "mw-category-group"})
    for element in elements:
        artistas.append(element.ul.li.a.text)
```

```python
artistas = artistas[4:]
```

```python
artistas_filtrados = []

for i in artistas:
    name = re.sub(r'\([^)]*\)|\[[^)]*\]', '', i) # remove (Deluxe edition) and [Feat. asdf] tags
    #name = re.sub(r'\W','', name).lower().strip() # remove all non-alphanumerical characters
    if ("Grupos" in name) or ("Músicos" in name):
        None
    else:
        artistas_filtrados.append(name)
```
